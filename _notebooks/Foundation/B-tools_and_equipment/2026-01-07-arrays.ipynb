## What are Arrays?



An **array** is a collection of elements (values) stored in a single variable. Each element in an array can be accessed by its **index** (position), starting from 0.

Think of an array like a row of boxes, each containing a value:
- Box 0 contains the first element
- Box 1 contains the second element  
- Box 2 contains the third element
- And so on...

### Why Use Arrays?
- Store multiple values in one variable
- Easy to access values by position
- Useful for loops (process many items at once)
- Organize related data together

```python
# Creating an array (list in Python)
fruits = ["apple", "banana", "orange", "grape"]

print("My array of fruits:")
print(fruits)
print()

# Accessing elements by index
print("First fruit (index 0):", fruits[0])
print("Third fruit (index 2):", fruits[2])
print()

# Array length
print("Number of fruits:", len(fruits))
```

## Modifying Arrays

You can change, add, or remove elements from an array:

```python
# Start with an array
colors = ["red", "blue", "green"]
print("Original array:", colors)
print()

# Change an element
colors[1] = "yellow"
print("After changing index 1 to 'yellow':", colors)
print()

# Add an element
colors.append("purple")
print("After appending 'purple':", colors)
print()

# Remove an element
colors.remove("red")
print("After removing 'red':", colors)
```

## Looping Through Arrays

Arrays are powerful when combined with loops. You can process each element without writing it by hand:

```python
# Using a for loop to go through each element
numbers = [10, 20, 30, 40, 50]

print("Numbers in the array:")
for num in numbers:
    print(num)

print()
print("Double each number:")
for num in numbers:
    print(num * 2)

print()
print("Using index to access elements:")
for i in range(len(numbers)):
    print(f"Index {i}: {numbers[i]}")
```

## Real-World Example: Student Grades

Let's calculate the average grade of students:

```python
# Array of student grades
grades = [85, 92, 78, 95, 88, 91]

# Calculate the sum
total = 0
for grade in grades:
    total = total + grade

# Calculate the average
average = total / len(grades)

print(f"Student grades: {grades}")
print(f"Total points: {total}")
print(f"Number of students: {len(grades)}")
print(f"Average grade: {average:.1f}")
print()

# Find the highest and lowest grades
print(f"Highest grade: {max(grades)}")
print(f"Lowest grade: {min(grades)}")
```
