## What are Classes? 
Imagine you are building a house. You don’t just start nailing boards together; you use a detailed set of plans. This blueprint defines all the characteristics a house should have: bedrooms, the size of kitchens, the number of bathrooms, and so on.

In JavaScript, we often need to create many similar “things”, like different user accounts, each with different properties, and different purposes they have. A JavaScript class is exactly like a blueprint, but for objects in code. It defines the properties (data), and methods (functions/actions). An object is like a standalone entity, with properties and type. For example, in a code file about cars, an object would be “car”.
